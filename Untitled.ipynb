```python
sport_and_number_url = {}

for img in image:
    values = img.split("/")[4]
    keys = img.split("/")[5]
    img
    sport_and_number_url.update({values:keys})
#     print(values)

sport_and_number_url
```

```python
urls = []
for k,v in sport_and_number_url.items():
    url = "https://bestof.one/categories/sports/articles/sport/"+ k +"/"+ v +""
    urls.append(url)
urls
```

```python
licensed_number = []
nom_alternatif = []
present_aux_jo = []
sport_olympique_depuis = []
type_de_sport = []
none = 'none'

for url_image in urls:
    print(url_image)
    cookies = dict(cookies_are ='working')
    req_second = requests.get(url_image, cookies = cookies)
    soup_second = BeautifulSoup(req_second.text,'html.parser')

    # Acces second_page
    corps = soup_second.find("div",class_ ="corps")
    secondtable =soup_second.findAll('section')[1]
    second_div = secondtable.article.findAll('div')[1]
    first_div = secondtable.article.findAll('div')[0]
    
    # Second table of second page NOM ALTERNATIF
    try:
        first_div.table.findAll('tr')[2]
        first_tr = first_div.table.findAll('tr')[2]
        second_td = first_tr.findAll('td')[1]
        nom_alternatif.append(second_td.string)
#         print('TRY',second_td.string)
    except :
        nom_alternatif.append('Pas de nom alternatif')
#         print('EXCEPT',second_td.string)
    
    
    # first table of second page LICENSED NUMBER
    second_tr = second_div.table.findAll('tr')[1]
    second_tr_td = second_tr.findAll('td')[1]
    # LICENSED NUMBER
    licensed_number.append(second_tr_td.string)
    
    
    
    # JEUX OLYMPIQUE
    try:
#         print(first_div.findAll('table')[1])
        table_jo = first_div.findAll('table')[1]
#         print(table_jo.tr.th.string)
        present_aux_jo.append(table_jo.tr.th.string)
        
    except:
        present_aux_jo.append('Non present au JO')
        pass
    
    
    
    
    # SPORT OLYMPIQUE DEPUIS...
    try:
#         print(first_div.findAll('table')[1])
        table_jo_2 = first_div.findAll('table')[1]
        table_jo_2_tr_2 = table_jo_2.findAll('tr')[1] 
#         print(table_jo_2_tr_2.td.string)
        table_jo_2_tr_2_td_2 = table_jo_2_tr_2.findAll('td')[1]
#         print(table_jo_2_tr_2_td_2.string)
        sport_olympique_depuis.append(table_jo_2_tr_2_td_2.string)
        
        
    except:
        sport_olympique_depuis.append('NaN')
        pass
    
    
    
    
    
    # TYPE DE SPORT OLYMPIQUE
    try:
        table_jo_3 = first_div.findAll('table')[1]
        table_jo_3_tr_2 = table_jo_3.findAll('tr')[2] 
#         print(table_jo_3_tr_2.td.string)
        table_jo_3_tr_2_td_2 = table_jo_3_tr_2.findAll('td')[1]
        print(table_jo_3_tr_2_td_2.string)
        type_de_sport.append(table_jo_3_tr_2_td_2.string)
        
    except:
        type_de_sport.append('Pas de présence qux jeux olympique')
        pass

# print(licensed_number)
# print(nom_alternatif)
# print(present_aux_jo)
print(type_de_sport)
```
